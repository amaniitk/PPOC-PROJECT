```python
import matplotlib.pyplot as plt
import numpy as np

# Define the features and their corresponding values
features = ['A', 'B', 'C']
values = [2, 3, 4]

# Define the colors for each feature
colors = ['red', 'blue', 'green']

# Create a bar chart using Matplotlib
plt.bar(features, values, color=colors)

# Add labels and title to the chart
plt.xlabel('Features')
plt.ylabel('Values')
plt.title('Bar Graph of Features A, B, and C')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the x and y coordinates of the points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot of the points
plt.scatter(x, y, color='red')

# Draw lines joining the points
for i in range(len(x)-1):
    plt.plot([x[i], x[i+1]], [y[i], y[i+1]], color='blue')

# Add labels and title to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cartesian Coordinates with Points and Lines')

# Display the plot
plt.show()
```
