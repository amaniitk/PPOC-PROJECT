```python
import matplotlib.pyplot as plt    ### Used mainly for data visualisation
```

```python
import numpy as np       ### importing the required libraries
```

```python
def sigmoid(z):
    return(1/(1+np.exp(-z)))   ### defining the sigmoid function
```

```python
def tanh(z):
    return((1-np.exp(-2*z))/(1+np.exp(-2*z)))    ### defining the tanh function
```

```python
#### Plotting the sigmoid function

X = np.linspace(-200, 200, 1000)    ### Sequence of 1000 numbers from -200 to 200
Y = sigmoid(X)
plt.plot(X, Y)                ### For plotting the graph of sigmoid
plt.xlabel("X")               ### Labels
plt.ylabel("Sigmoid(X)")
plt.title("Sigmoid function")

plt.show()                    ### Showing the plot
```

```python
#### Plotting the tanh function

X = np.linspace(-100, 100, 1000)    ### Sequence of 1000 numbers from -200 to 200
Y = tanh(X)
plt.plot(X, Y)                ### For plotting the graph of hyperbolic tangent function
plt.xlabel("X")               ### Labels
plt.ylabel("tanh(X)")
plt.title("Hyperbolic tangent")

plt.show()                    ### Showing the plot
```
